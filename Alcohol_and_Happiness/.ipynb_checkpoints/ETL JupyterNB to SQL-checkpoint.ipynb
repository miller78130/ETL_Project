```python
import numpy as np
import matplotlib as mp
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
import scipy.stats as st
```

```python
# Set up for the scatter plot
plt.figure(figsize = (15,5))
sns.scatterplot(data = Joins_NewCut_df, x = "healthy_life_expectancy", y = "happinessscore", palette = "turbo")

# Set x and y values
x_values = Joins_NewCut_df["healthy_life_expectancy"]
y_values = Joins_NewCut_df["happinessscore"]

# Set values for plot
plt.title("Happiness Score and Healthy Life Expectancy for all Countries", fontsize = 20)
plt.xlabel("Healthy Life Expectancy", fontsize = 15)
plt.ylabel("Happiness Score", fontsize = 15)
plt.savefig("Figures/hp_le_scatter.png")
plt.show()
```
