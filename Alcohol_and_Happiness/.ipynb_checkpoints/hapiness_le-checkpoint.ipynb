```python
# Load Dependacies
import pandas as pd
import numpy as np
import matplotlib as mp
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
import scipy.stats as st
```

```python
# Load read of Hapiness Score and Length of Life csv join
hp_le_csv = "Resources/Joins HP_LE_WH.csv"

# Read CSV
hp_le_df = pd.read_csv(hp_le_csv)
hp_le_df.head()
```

```python
# Set order by happiness score
hp_le_sort_df = hp_le_df.sort_values("happinessscore", ascending = False)
hp_le_sort_df.head()
```

```python
# Set up for the scatter plot
plt.figure(figsize = (15,5))
sns.scatterplot(data = hp_le_df, x = "healthy_life_expectancy", y = "happinessscore", palette = "turbo")

# Set x and y values
x_values = hp_le_df["healthy_life_expectancy"]
y_values = hp_le_df["happinessscore"]

# Set values for plot
plt.title("Happiness Score and Healthy Life Expectancy for all Countries", fontsize = 20)
plt.xlabel("Healthy Life Expectancy", fontsize = 15)
plt.ylabel("Happiness Score", fontsize = 15)
plt.savefig("Figures/hp_le_scatter.png")
plt.show()
```
